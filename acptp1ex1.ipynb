```r
# Paramétrage
n = 30
library(ggplot2)
```

```r
# Génération de loi normales ordonnées en 2 colonnes
set.seed(123)
x = matrix(rnorm(n * 3 * 2), n * 3, 2)
for (i in 31:60) x[i,] <- x[i,] + c(10, 0)
for (i in 61:90) x[i,] <- x[i,] + c(0, 10)
colnames(x) <- c("X1", "X2")
x
```

```r
x = as.data.frame(x)
ggplot(x,aes(x=X1,y=X2)) + geom_point()
```

```r
x = scale(x)
x=as.data.frame(x)
ggplot(x,aes(x=X1,y=X2)) + geom_point()
```

```r
cl = as.factor(c(rep(1,n), rep(2,n), rep(3,n)))
cl
```

```r
ggplot(x,aes(x=X1,y=X2,colour=cl)) + geom_point()
```
